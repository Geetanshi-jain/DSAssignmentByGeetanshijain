```python
#k-Means Clustering


# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs


# Generate synthetic data for demonstration

# We generate 300 samples with 3 clusters
X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
"""
n_samples=300: Creates 300 data points.
centers=3: The data points are grouped into 3 distinct clusters.
cluster_std=0.60: Determines the spread of the clusters.
random_state=0: Ensures reproducibility of the random numbers (same data each time)."""

"""
X: The features (data points).
y: The actual cluster labels (for our knowledge, but not used in k-Means)."""
# Create a DataFrame
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])

#Fitting k-Means:
# Apply k-Means clustering
"""Here, we apply the k-Means algorithm:

n_clusters=3: Tells k-Means to create 3 clusters (as we generated 3 in our data).
random_state=42: Ensures the same centroids initialization every time for reproducibility.
fit(df): Applies the k-Means algorithm on the DataFrame."""

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df)
"""Once k-Means has fitted the data, each data point is assigned to one of the clusters.
 The labels (kmeans.labels_) indicate which cluster each point belongs to.
\We add these labels to the DataFrame as a new column named 'Cluster'."""
# Get cluster labels and cluster centers
df['Cluster'] = kmeans.labels_

#Extracting Cluster Centers:
centers = kmeans.cluster_centers_


# Visualize the clusters and their centers
plt.figure(figsize=(8, 6))
plt.scatter(df['Feature1'], df['Feature2'], c=df['Cluster'], cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], s=300, c='red', label='Centroids', marker='X')
plt.title('k-Means Clustering')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.legend()
plt.show()
```
