```python
#1. Construct a Bar Graph with Overlay Using Python
# Bar Graph -A bar graph is a chart that uses rectangular bars to represent and compare different categories or groups. The length or height of each bar reflects
 #the value or frequency of the category. For example, a bar graph can show the number of books read by different students.
import numpy as np
import matplotlib.pyplot as plt

# Example data
categories = ['A', 'B', 'C', 'D']
count1 = [10, 20, 15, 25]
count2 = [12, 18, 22, 19]

x = np.arange(len(categories))  # Creating positions for the categories
width = 0.35  # Width of the bars

# Plotting bars with overlay
fig, ax = plt.subplots()
ax.bar(x - width/2, count1, width, label='Count1', alpha=0.7)  # Bar 1
ax.bar(x + width/2, count2, width, label='Count2', alpha=0.7)  # Bar 2

ax.set_xticks(x)  # Setting x-ticks
ax.set_xticklabels(categories)  # Labeling categories on the x-axis
ax.set_title('Bar Graph with Overlay')  # Graph title
ax.legend()  # Adding legend

plt.show()

```

```python
# Construct Contingency Tables Using Python
import pandas as pd

# Example data
data = {'Gender': ['Male', 'Female', 'Female', 'Male', 'Male'],
        'Choice': ['Yes', 'No', 'Yes', 'No', 'Yes']}

df = pd.DataFrame(data)

# Create contingency table
contingency_table = pd.crosstab(df['Gender'], df['Choice'])
print(contingency_table)

```

```python
#A histogram is a type of bar chart that shows how data is spread out or distributed. It divides the
#data into ranges (called "bins") and then tells you how many values fall into each range.
 #For example, if you're looking at people's ages, a histogram can show you how many people fall into different age groups, like 10-20 years, 20-30 years, a
 #nd so on. This helps you see the shape of the data, such as if most people are younger or older.
import numpy as np
import matplotlib.pyplot as plt

# Example data
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(55, 12, 100)

# Plot histograms with overlay
plt.hist(data1, bins=15, alpha=0.5, label='Data1', color='blue')  # Histogram 1
plt.hist(data2, bins=15, alpha=0.5, label='Data2', color='red')   # Histogram 2

plt.title('Histogram with Overlay')  # Title of the graph
plt.legend()  # Display legend for the two datasets
plt.show()  # Show the plot


```

```python
#4.Perform Binning Based on Predictive Value Using Python
import pandas as pd

# Example data
data = {'Age': [22, 25, 35, 40, 50],
        'Income': [30000, 40000, 50000, 55000, 70000]}

df = pd.DataFrame(data)

# Perform binning
df['AgeGroup'] = pd.cut(df['Age'], bins=[20, 30, 40, 50], labels=['20-30', '30-40', '40-50'])

print(df)


```
