```python
import matplotlib.pyplot as plt

# Combined performance data
categories = [
    'Class 5', 'Class 6', 'Class 7', 'Class 8',
    'Class 9', 'Class 10', 'Class 11', 'Class 12',
    'Semester 1', 'Semester 2', 'Semester 3', 'Semester 4',
]
scores = [80, 82, 85, 88, 91, 95, 89, 91, 92, 90, 77, 88, ]

# Create the bar graph
plt.figure(figsize=(14, 6))
plt.bar(categories, scores, color='lightblue', edgecolor='black')

# Add labels and title
plt.title('Performance from Class 5 to Semester 5', fontsize=16)
plt.xlabel('Categories', fontsize=14)
plt.ylabel('Scores (%)', fontsize=14)
plt.xticks(rotation=45, fontsize=10)  # Rotate category labels for better readability

# Show scores above each bar
for i in range(len(scores)):
    plt.text(i, scores[i] + 1, str(scores[i]), ha='center', fontsize=10)

# Display the graph
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
