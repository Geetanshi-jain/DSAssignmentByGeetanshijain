```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "sensor_data.csv"  # Replace with your file path
df = pd.read_csv("/content/drive/MyDrive/Mca/hydrowatch.csv")

# Convert the Timestamp column to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Set the Timestamp as the index
df.set_index('Timestamp', inplace=True)

# Plotting
plt.figure(figsize=(12, 8))

# Turbidity plot
plt.subplot(2, 2, 1)
plt.plot(df.index, df['Turbidity (NTU)'], marker='o', color='blue', label='Turbidity (NTU)')
plt.title('Turbidity Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Turbidity (NTU)')
plt.grid(True)
plt.legend()

# Flow Rate plot
plt.subplot(2, 2, 2)
plt.plot(df.index, df['Flow Rate (ml/sec)'], marker='s', color='green', label='Flow Rate (ml/sec)')
plt.title('Flow Rate Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Flow Rate (ml/sec)')
plt.grid(True)
plt.legend()

# Volume plot
plt.subplot(2, 2, 3)
plt.plot(df.index, df['Volume (ml)'], marker='^', color='orange', label='Volume (ml)')
plt.title('Volume Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Volume (ml)')
plt.grid(True)
plt.legend()

# Water Level plot
plt.subplot(2, 2, 4)
plt.plot(df.index, df['Water Level (0-1024)'], marker='d', color='red', label='Water Level')
plt.title('Water Level Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Water Level (0-1024)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
